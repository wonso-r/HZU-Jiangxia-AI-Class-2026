```python
import cv2
import numpy as np
import math

# 展示图像函数
def show_image(image, title='Image'):
    cv2.imshow(title, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# 读取图像
image = cv2.imread('./wlsj/pm2(1).jpg')

# 检查图像是否成功加载
if image is None:
    print("图像加载失败，请检查路径。")
else:
    # resize
    resized_image = cv2.resize(image, (0, 0), fx=0.5, fy=0.5)
    show_image(resized_image)

    # 转换为灰度图
    gray = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
    show_image(gray)

    # 使用更温和的去噪方式，比如中值滤波，降低对小圆形边缘的影响
    blurred = cv2.medianBlur(gray, 3)
    show_image(blurred)

    # 使用Canny边缘检测
    edges = cv2.Canny(blurred, 30, 90)
    show_image(edges)

    # 精细调整HoughCircles参数以检测最小的圆
    circles = cv2.HoughCircles(edges, cv2.HOUGH_GRADIENT, dp=0.3, minDist=1.5,
                               param1=6, param2=13, minRadius=1, maxRadius=8)

    # 确保检测到圆
    if circles is not None:
        # 将圆检测结果从浮点数转换为整数
        circles = np.round(circles[0, :]).astype("int")

        # 获取两个圆的圆心坐标（假设检测到了两个圆，如果实际情况不同需调整逻辑）
        circle1_x, circle1_y, _ = circles[0]
        circle2_x, circle2_y, _ = circles[1]

        # 计算两圆连线的斜率
        if circle2_x - circle1_x == 0:  # 防止斜率无穷大的情况（两圆横坐标相同）
            slope = float('inf')
        else:
            slope = (circle2_y - circle1_y) / (circle2_x - circle1_x)

        # 根据斜率计算与水平方向的夹角（弧度制）
        angle_rad = math.atan(slope)

        # 将弧度制转换为角度制
        angle_deg = angle_rad * 180 / math.pi

        print("两圆连线与水平方向的夹角为:", angle_deg, "度")

        # 在图像上画出所有检测到的圆，这里考虑更准确的还原到原始图像尺寸的坐标和半径计算
        for (x, y, r) in circles:
            # 由于图像被缩小了，所以需要将坐标和半径放大回原始尺寸，根据缩放比例计算还原
            origin_x = int(x / 0.5)
            origin_y = int(y / 0.5)
            origin_r = int(r / 0.5)
            # cv2.circle(image, (origin_x, origin_y), origin_r, (0, 255, 0), 2)

        # 点出圆心，不同颜色
        cv2.circle(image, (int(circle1_x / 0.5), int(circle1_y / 0.5)), 4, (255, 0, 0), -1)
        cv2.circle(image, (int(circle2_x / 0.5), int(circle2_y / 0.5)), 4, (0, 255, 0), -1)

        # 绘制两圆连线，还原到原始图像尺寸坐标
        origin_circle1_x = int(circle1_x / 0.5)
        origin_circle1_y = int(circle1_y / 0.5)
        origin_circle2_x = int(circle2_x / 0.5)
        origin_circle2_y = int(circle2_y / 0.5)
        cv2.line(image, (origin_circle1_x, origin_circle1_y), (origin_circle2_x, origin_circle2_y), (0, 0, 255), 2)

        # 显示结果图像
        show_image(image, 'Detected Circles')
    else:
        print("没有检测到圆。")

# 保存图像到文件
cv2.imwrite('detected_circles3.jpg', image)
```

```python
import cv2
import numpy as np

def show_image(image, title='Image'):
    cv2.imshow(title, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# 读取图像
image = cv2.imread('./wlsj/VSLOT2(1).jpg')

# 检查图像是否成功加载
if image is None:
    print("图像加载失败，请检查路径。")
else:
    # 转换为灰度图
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    show_image(gray, 'gray')

    # 应用高斯模糊来减少噪声
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    show_image(blurred, 'blurred')

    # 使用Canny边缘检测
    edges = cv2.Canny(blurred, 30, 100)
    show_image(edges, 'edges')

    # 找到边缘的轮廓
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # 找到最大的轮廓，假设它是零件的轮廓
    max_contour = max(contours, key=cv2.contourArea)

    # 创建掩模
    mask = np.zeros_like(gray)  # 使用与灰度图相同大小和类型的掩模
    # 掩膜的作用是只保留我们感兴趣的区域，即零件的轮廓

    # 在掩模上绘制零件轮廓
    cv2.drawContours(mask, [max_contour], -1, (255), thickness=cv2.FILLED)

    # 应用掩模
    masked_edges = cv2.bitwise_and(edges, edges, mask=mask)
    show_image(masked_edges, 'Masked Edges')

    # 找到掩模边缘的轮廓
    masked_contours, _ = cv2.findContours(masked_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # 存储槽点
    slot_points = []

    # 如果存在轮廓
    if masked_contours:
        for contour in masked_contours:
            # 计算轮廓的凹点
            hull = cv2.convexHull(contour, returnPoints=False)
            defects = cv2.convexityDefects(contour, hull) 
            # cv2.convexityDefects(contour, hull)表示计算轮廓的凹点，返回一个包含凹点信息的数组
            if defects is not None:
                for i in range(defects.shape[0]):
                    s, e, f, d = defects[i, 0]
                    start = tuple(contour[s][0])
                    end = tuple(contour[e][0])
                    far = tuple(contour[f][0])
                    if d > 3000:  # 使用适当的距离阈值
                        slot_points.append(far)
                        cv2.circle(image, far, 5, (0, 0, 255), -1)

    # 可视化结果
    cv2.drawContours(image, masked_contours, -1, (0, 255, 0), 2)
    cv2.imshow('Detected V-Notches', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    # 打印槽点坐标
    for point in slot_points:
        print(point)
```

```python
import cv2
import numpy as np

def show_image(image, title='Image'):
    cv2.imshow(title, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# 读取图像
image = cv2.imread('./wlsj/VSLOT2(1).jpg')

# 检查图像是否成功加载
if image is None:
    print("图像加载失败，请检查路径。")
else:
    # 转换为灰度图
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # 应用高斯模糊来减少噪声
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    
    # 使用Canny边缘检测
    edges = cv2.Canny(blurred, 40, 100)  # 尝试不同的阈值
    
    # 找到边缘的轮廓
    contours, _ = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # 找到最大的轮廓，假设它是零件的轮廓
    max_contour = max(contours, key=cv2.contourArea)
    
    # 创建掩模
    mask = np.zeros_like(gray)
    cv2.drawContours(mask, [max_contour], -1, (255), thickness=cv2.FILLED)
    
    # 应用掩模
    masked_edges = cv2.bitwise_and(edges, edges, mask=mask)
    show_image(masked_edges, 'Masked Edges')
    
    # 找到掩模边缘的轮廓
    masked_contours, _ = cv2.findContours(masked_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # 存储槽点
    slot_points = []
    
    # 如果存在轮廓
    if masked_contours:
        for contour in masked_contours:
            # 计算轮廓的凹点
            hull = cv2.convexHull(contour, returnPoints=False) 
            # cv2.convexHull(contour, returnPoints=False)表示计算轮廓的凸包，返回一个包含凸包点的数组
            defects = cv2.convexityDefects(contour, hull)
            # cv2.convexityDefects(contour, hull)表示计算轮廓的凹点，返回一个包含凹点信息的数组
            
            if defects is not None:
                for i in range(defects.shape[0]):
                    s, e, f, d = defects[i, 0]
                    start = tuple(contour[s][0])
                    end = tuple(contour[e][0])
                    far = tuple(contour[f][0])
                    
                    # 增加额外的过滤条件，如角度和距离
                    angle = abs(np.math.degrees(np.arctan2(start[1] - end[1], start[0] - end[0])))
                    distance = np.sqrt((start[0] - end[0])**2 + (start[1] - end[1])**2)
                    
                    if d > 3000 and angle < 90 and distance > 10:  # 根据实际情况调整阈值
                        slot_points.append(far)
                        cv2.circle(image, far, 5, (0, 0, 255), -1)
    
    # 可视化结果
    cv2.drawContours(image, masked_contours, -1, (0, 255, 0), 2)
    cv2.imshow('Detected V-Notches', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    # 打印槽点坐标
    for point in slot_points:
        print(point)
```

```python
import cv2
import numpy as np
import math

def show_image(image, title='Image'):
    cv2.imshow(title, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# 读取图像
image = cv2.imread('./wlsj/VSLOT2(1).jpg')

# 检查图像是否成功加载
if image is None:
    print("图像加载失败，请检查路径。")
else:
    # 转换为灰度图
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    show_image(gray, 'Gray Image')

    # 增强对比度，进行对比度拉伸
    gray = cv2.convertScaleAbs(gray, alpha=1.5, beta=0)
    # 各个参数的含义：cv2.convertScaleAbs(gray, alpha=1.5, beta=0)表示对图像进行对比度拉伸，alpha表示对比度因子，beta表示亮度因子。
    # alpha越大，对比度越大；beta越大，亮度越大。
    show_image(gray, 'Contrast Enhanced Image')

    # 应用高斯模糊，减少噪声
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    show_image(blurred, 'Blurred Image')
    
    # 自适应二值化处理
    binary = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15, 6)
    # 各个参数的含义：cv2.ADAPTIVE_THRESH_GAUSSIAN_C表示使用高斯窗口，cv2.THRESH_BINARY_INV表示反二值化，11表示窗口大小，2表示C值，c用于调整阈
    # 窗口大小越大，二值化效果越好，但可能会丢失细节；C值越大，二值化效果越差，但可能会保留更多细节。
    show_image(binary, 'Binary Image')
    
    # 形态学操作：开运算去除噪声
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
    morph = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel, iterations=2)
    
    # 再次使用膨胀操作使槽点区域更明显
    morph = cv2.dilate(morph, kernel, iterations=2)
    
    # 查找轮廓
    contours, _ = cv2.findContours(morph, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # 存储槽点坐标
    slot_points = []
    
    # 遍历所有轮廓
    for contour in contours:
        # 过滤轮廓，根据面积和形状筛选出槽点
        area = cv2.contourArea(contour)
        if area > 50:  # 根据实际情况调整面积阈值
            # 获取轮廓的最小外接矩形
            rect = cv2.minAreaRect(contour)
            box = cv2.boxPoints(rect)
            box = np.int0(box)
            
            # 计算槽点的中心位置
            cx = int(rect[0][0])
            cy = int(rect[0][1])
            slot_points.append((cx, cy))
            
            # 绘制轮廓和中心点
            # cv2.drawContours(image, [box], 0, (0, 255, 0), 2)
            cv2.circle(image, (cx, cy), 5, (0, 0, 255), -1)
    
    # 按X坐标对槽点排序（假设槽点从左到右排列）
    slot_points = sorted(slot_points, key=lambda x: x[0])
    
    # 计算槽点之间的距离、角度和深度
    if len(slot_points) > 1:
        for i in range(1, len(slot_points)):
            prev_point = slot_points[i - 1]
            current_point = slot_points[i]
            
            # 计算槽点间的距离
            distance = math.sqrt((current_point[0] - prev_point[0])**2 + (current_point[1] - prev_point[1])**2)
            
            # 计算槽点间的角度
            angle = math.degrees(math.atan2(current_point[1] - prev_point[1], current_point[0] - prev_point[0]))
            
            # 假设槽深度为槽点到零件底部的距离（根据实际情况调整）
            depth = abs(current_point[1] - max([point[1] for point in slot_points]))
            
            # print(f"槽点 {i}: 距离 = {distance:.2f}，角度 = {angle:.2f}°，深度 = {depth:.2f}")
            
            # 在图像上显示槽点的连接线和标注
            # cv2.line(image, prev_point, current_point, (255, 0, 0), 1)
            # cv2.putText(image, f"D:{distance:.2f}", (prev_point[0], prev_point[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255, 255, 255), 1)
            # cv2.putText(image, f"A:{angle:.2f}", (prev_point[0], prev_point[1] - 25), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255, 255, 255), 1)
            # cv2.putText(image, f"H:{depth:.2f}", (prev_point[0], prev_point[1] - 40), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255, 255, 255), 1)
    
    # 显示结果
    show_image(image, 'Detected Slots with Distances, Angles, and Depths')
    
    # # 打印槽点坐标
    # for i, point in enumerate(slot_points):
    #     print(f"槽点 {i+1} 坐标: {point}")
```
