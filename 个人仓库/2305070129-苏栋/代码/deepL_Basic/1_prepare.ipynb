自动求梯度

```python
from mxnet import autograd,nd
```

```python
x=nd.arange(4).reshape(4,1)
x
```

```python
x.attach_grad()
```

```python
with autograd.record():
    y=2*nd.dot(x.T,x)
```

```python
y.backward()
```

```python
assert(x.grad-4*x).norm().asscalar()==0
```

4 * x：

这是预期的梯度值。假设你有一个损失函数，其梯度计算结果是 4 * x。

x.grad - 4 * x：

计算实际梯度与预期梯度之间的差值。

.norm()：

计算差值的范数（默认是 L2 范数）。范数表示差值的大小。

.asscalar()：

将范数从 mxnet.ndarray.NDArray 类型转换为 Python 标量（float 或 int）。

```python
x.grad
```

梯度是损失函数对模型参数的偏导数，它指示了参数应该如何更新以最小化损失函数。
