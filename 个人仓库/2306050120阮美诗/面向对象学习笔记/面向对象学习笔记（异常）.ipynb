抛出异常：程序执行时遇到错误，停止执行，并提示错误信息。

通过异常捕获可以针对异常事件集中处理，保证程序面面俱到。

对某些代码的执行不能确定是否正确可以用 try 来表达异常

```python
try: #下面是不确定是否正确运行的代码
    num = int(input("请输入一个整数："))
except:
    print("请输入正确的整数")
print("-"*50)
```

捕获错误类型:抛出错误信息的最后一行的第一个单词就是错误类型

```python
num = int(input("请输入一个整数："))
print(num) #可见异常类型是值异常（ValueError）
```

```python
#捕获已知错误
try:
    num = int(input("请输入一个整数："))
    result = 8 / num
    print(result)

except ZeroDivisionError:
    print("除0错误")
except ValueError:
    print("值错误")
```

```python
#当我们不能预知所有错误类型时，需要识别未知的错误类型
try:
    num = int(input("请输入一个整数："))
    result = 8 / num
    print(result)

except ZeroDivisionError:
    print("除0错误")
except Exception as result: #从异常库中识别错误类型
    print("未知错误 %s " % result)
```

异常传递：当函数/方法执行出现异常，程序不会被立即终止，会将异常传递给调用的一方；
直到传递给主程序，仍有没有处理才会终止

```python
def now():
    num = int(input("请输入一个整数："))
def now2():
    now()
print(now2())
#可以看到报错从now一直传递到最后print
```

```python
def demo1():
    return int(input("输入整数："))
def demo2():
    return demo1()
try:
    print(demo2())
except Exception as result:
    print("未知错误%s" % result)
#因为异常的传递性，只需要在最后主函数进行异常处理就行
```

这里总结了一个完整全面的异常捕获语法

```python
try: #添加可能无法正确运行的代码
    num = int(input("请输入一个整数："))
    result = 8 / num
    print(result)

except ZeroDivisionError:
    print("除0错误") #识别已知异常
except Exception as result: #从异常库中识别错误类型
    print("未知错误 %s " % result) #检测未知异常
else:
    print("没有异常时执行")
finally:
    print("不管有没有异常都执行")
```

```python
除了因为执行出错系统抛出的异常，当业务有特定的需求时可以主动抛出异常
```

```python
def input_password():
    pwd = input("请输入密码：")
    if len(pwd) == 8:
        return pwd
    
    ex = Exception("密码长度不够") #主动抛出的步骤 1、在异常库添加自定义的异常类型
    raise ex #步骤 2、自定义异常类型的抛出
try:
    print(input_password())
except Exception as result:
    print(result) 
```
