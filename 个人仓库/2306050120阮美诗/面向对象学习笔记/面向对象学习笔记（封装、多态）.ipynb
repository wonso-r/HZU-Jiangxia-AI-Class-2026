封装

```python
class Person(object):
    def __init__(self,name,age):
        self.name = name
        self.age = age
        self.life_val = 100
a = Person("Alex",22)
a.life_val -= 50
print(a.life_val)#可以看到定义在类中的属性生命值可以在类外被修改，要防止这个情况出现就需要封装
```

```python
class Person(object):
    def __init__(self,name,age):
        self.name = name
        self.age = age
        self.__life_val = 100#加两个下划线，完成封装
    
    def cut_life_val(self):
        self.__life_val -= 20#封装后只能通过这个方法修改被封装后的属性
        print(self.__life_val)
        self.__breath()
    
    def __breath(self):
        print("%s is breating "%self.name)

a = Person("Alex",22)
a.cut_life_val()
```

```python
a.__breath()#被封装后的方法不能在类的外面调用
#可以通过特殊方法访问
```

```python
a._Person__breath()#实例+_类名+方法名，可以访问被封装的方法
a._Person__life_val -= 20 
print(a._Person__life_val)
```

多态

```python
class Dog(object):
    def sound(self):
        print("wang")

class Cat(object):
    def sound(self):
        print("meow")

def make_sound(self):
    self.sound()

d1 = Dog()
c1 = Cat()
#这个例子是通过函数实现的多态现象，调用同一个函数接口，（区别名称种类）输出不同的函数效果
make_sound(d1)
make_sound(c1)
```

```python
#用抽象类实现多态则更为常见
class Document:
    def __init__(self,name):
        self.name = name
    
    def show(self):
        raise NotImplementedError("Subclass must implement abstract method")
        #手动报错，强制让show函数被改写，否则异常

class Pdf(Document):
    def show(self):
        return "show pdf contents"

class Word(Document):
    def show(self):
        return "show word contents"

p = Pdf("pdf")
p.show()#本质上还是不同的接口，调用时还区分p/w

objs = [p,w]
for i in objs:
    i.show#现在归类成一个接口
```
