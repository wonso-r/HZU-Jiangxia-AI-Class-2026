```python
#hasattr
class Person():
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
p = Person("Alex",22)
if hasattr(p,"name2"): #如果程序员只获得这个类的实例p，而没有获取这个类的定义，ta不知道这个属性名是什么可能是是name，可能是name1。。。
    print("have")
#使用 hasatter 全称是 "Has Attribute"（拥有属性）， 是一种反射（自省）的方法，可以判断实例p是否拥有这个“name2”这个属性
#引用这个检查可以检查属性是否存在的同时防止报错崩溃
```

反射对应四种方法：增（setattr）、删（delattr）、改（getattr）、查（hasattr）


```python
#getattr
class Person():
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def walk(self):
        print("walking...")
        
p = Person("Alex",22)
a = getattr(p,"age")
print(a)

#该方法可以通过 字符串 的形式获取实例的属性值，也可以把hasattr结合起
#来，判断用户输入的字符串检查属性并获取对应值
user_command = input(">>:").strip()
if hasattr(p,user_command): #这里输错不报错，输入正确通过检测即可获取属性值，全程字符串检测
    func = getattr(p,user_command)
    func()
```

```python
#setattr
class Person():
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def walk(self):
        print("walking...")

def talk(self): #这是一个定义在类外的函数，需要绑定在类中
    print(self.name , "is speaking ...")
        
p = Person("Alex",22)
setattr(p,"gender","F") #给实例设置静态属性，（实例名 + 属性名 + 属性值）
print(p.gender)

setattr(p,"speak",talk)#可以通过这个办法将类外的函数绑定为该实例的一个方法,（属性 + 属性名（自定义）+ 函数名）
p.speak(p)
setattr(Person,"speak2",talk)
p.speak2()
```

```python
#运用映射方式
class User:
    def login(self):
        print("欢迎来到登录页面")
    def register(self):
        print("欢迎来到注册页面")
    def save(self):
        print("保存")

user = User()
while 1:
    choose = input('>>:').strip()
    if hasattr(user,choose):
        func = getattr(user,choose)
        func()
    else:
        print("输入错误")
```
